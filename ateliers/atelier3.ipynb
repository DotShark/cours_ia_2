### Descente de gradient

PyTorch calcule automatiquement les gradients via `autograd`. En activant `requires_grad=True` sur un tenseur et en appelant `.backward()`, on obtient la dérivée de la fonction par rapport à ce tenseur.

Ici, pour `y = x² + 2x + 1` en `x = 3`, la dérivée `2x + 2 = 8` est bien retrouvée automatiquement.

```python
x = torch.tensor(3.0, requires_grad=True)
y = x**2 + 2*x + 1
y.backward()
```

```python
print("\n[1.4] x =", x.item(), "-> y = x²+2x+11 =", y.item())
print("  x.grad (calculé par autograd)", x.grad.item())
print("  dérivée analytique 2x+2 pour x=3 =", 2*3+2, "✓" if x.grad.item() == 8 else "x")
```

### Perceptron

Un perceptron simple avec une couche linéaire `nn.Linear(784, 10)` : 784 entrées (images 28×28 pixels de MNIST) et 10 sorties (une par classe de chiffre).

```python
from torch import nn

class Perceptron(nn.Module):
    def __init__(self):
        super().__init__()
        self.perceptron = nn.Linear(28*28, 10)
    
    def forward(self, x):
        y = self.perceptron(x)
        return y
```

```python
perceptron = Perceptron()
next(perceptron.parameters()).shape
```

```python
x_dummy = torch.randn(10, 784)
output = perceptron(x_dummy)
output.shape
```
